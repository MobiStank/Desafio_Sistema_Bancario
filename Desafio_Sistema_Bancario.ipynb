```python
from time import sleep

dinheiro_disponivel = 2000
dinheiro_na_conta = 0
extratos_dos_depositos = []
extratos_dos_saques = []
TENTATIVAS = [1,2,3]

while True:
    print('\033[32mBanco Desafio\033[m'.center(30, '='))
    opção = int(input('\033[34mEscolha o que deseja fazer:\n(1)Depositar\n(2)Sacar\n(3)Tirar extrato\n(4)Sair\033[m\n\033[32mEscolha: \033[m'))
    sleep(0.5)

    if opção == 1:
        print('\033[32mDepositar\033[m'.center(30, '='))
        sleep(0.1)
        valor_do_deposito = float(input('\033[34mDeseja depositar quanto? R$\033[m'))

        continuar = int(input('\033[34mDeseja continuar?\n(1)sim\n(2)não\033[m\n\033[32mEscolha: \033[m'))
        print('\n')

        if valor_do_deposito <= dinheiro_disponivel and continuar == 1:
            dinheiro_na_conta += valor_do_deposito
            extratos_dos_depositos += [f'Desposito {len(extratos_dos_depositos) + 1}: R${valor_do_deposito:.2f}']
            print('\033[33mAnalisando...\033[m')
            sleep(1)
            print('\033[33mInsira o Dinheiro!\033[m')
            sleep(1)
            print('\033[33mDespositando...\033[m')
            sleep(1)
            print('\033[32mDespositado!\033[m\n')
            sleep(0.5)
        elif valor_do_deposito > dinheiro_disponivel:
            print('\033[33mAnalisando...\033[m')
            sleep(1)
            print('\033[31mFalha! Não foi possivel depositar o valor, verifique se o valor do dinheiro é maior ou igual ao valor do deposito.\033[m')
            sleep(0.5)
        elif continuar == 2:
            print('\033[33mVoltando...\033[m')
            sleep(1)
            continue

        sair = int(input('\033[34mDeseja Voltar para o menu?\n(1)sim\n(2)não(fechar)\033[m\n\033[32mEscolha: \033[m'))
        print('\n')
        
        if sair == 1:
            print('\033[33mVoltando...\033[m\n')
            sleep(1)
            continue
        elif sair == 2:
            exit()
        else:
            print('\033[31mResposta Inválida\033[m')
            continue
    
    elif opção == 2:
        print('\033[32mSacar\033[m'.center(30, '='))
        sleep(0.1)
        valor_do_saque = float(input('\033[34mQuanto deseja sacar\033[m\033[31m(MAX: R$500)\033[m: \033[32mR$\033[m'))
        print('\n')

        continuar = int(input('\033[34mDeseja continuar?\n(1)sim\n(2)não\033[m\n\033[32mEscolha: \033[m'))
        print('\n')
        
        if valor_do_saque <= dinheiro_na_conta and valor_do_saque <= 500 and continuar == 1:
            
            if len(TENTATIVAS) == 3 or len(TENTATIVAS) == 2 or len(TENTATIVAS) == 1:
                del TENTATIVAS[-1]
                dinheiro_na_conta -= valor_do_saque
                extratos_dos_saques += [f'Saque {len(extratos_dos_saques) + 1}: R${valor_do_saque:.2f}']
                print('\033[33mAnalisando...\033[m')
                sleep(1)
                print(f'\033[32mSucesso! Sacando o valor de R${valor_do_saque:.2f}\033[m.\n')
                sleep(0.5)
            elif len(TENTATIVAS) == 0:
                sleep(0.7)
                print('\033[31mSaques maximo diario execido! Tente amanhã\033[m')
        
        elif (valor_do_saque > dinheiro_na_conta or valor_do_saque > 500) and continuar == 1:
            print('\033[33mAnalisando...\033[m')
            sleep(1)
            print('\033[31mFalha! Não foi possivel sacar este valor, verifique o dinheiro na conta ou verifique se o valor do saque excedi o valor maximo diario de R$500.\033[m')
            sleep(0.5)
        elif continuar == 2:
            print('\033[33mVoltando...\033[m')
            sleep(1)
            continue

        sair = int(input('\033[34mDeseja Voltar para o menu?\n(1)sim\n(2)não(fechar)\033[m\n\033[32mEscolha: \033[m'))
        print('\n')

        if sair == 1:
            print('\033[33mVoltando...\033[m\n')
            sleep(1)
            continue
        elif sair == 2:
            exit()
    
    elif opção == 3:
        print('\033[32mTirar Extrato\033[m'.center(30, '='))
        escolha = int(input('\033[34mDeseja tirar o extrato da conta?\n(1)sim\n(2)não\033[m\n\033[32mEscolha: \033[m'))
        print('\n')

        if escolha == 1:
            print('\033[33mCriando extrato da conta...\033[m')
            sleep(1)
            print('\033[32mExtrato Criado!\033[m\n')
            sleep(1)
            visualizar_extrato = int(input('\033[34mDeseja visualizar o extrato da conta?\n(1)sim\n(2)não\033[m\n\033[32mEscolha: \033[m'))
            print('\n')

            if visualizar_extrato == 1:
                print('\033[33mAguarde um momento...\033[m\n')
                sleep(1)
                print('\033[32mExtrato\033[m'.center(49, '='))

                try:
                    print(extratos_dos_saques[0])
                except:
                    pass
                try:
                    print(extratos_dos_saques[1])
                except:
                    pass
                try:
                    print(extratos_dos_saques[2])
                except:
                    pass
                
                print('\n')

                try:
                    print(extratos_dos_depositos[0])
                except:
                    pass
                try:
                    print(extratos_dos_depositos[1])
                except:
                    pass
                try:
                    print(extratos_dos_depositos[2])
                except:
                    pass
                    
                print(f'\n\nDinheiro na Conta:            R${dinheiro_na_conta:.2f}')
                print(f'\nDuvidas?\nEntre em contato conosco: (55) 4747-4333\n{"="*41}\n')
                sleep(1)
            elif visualizar_extrato == 2:
                print('\033[33mVoltando...\033[m')
                sleep(1)
                continue
        
        elif escolha == 2:
            print('\033[33mVoltando ...\033[m')
            sleep(1)
            continue

    elif opção == 4:
        print('\033[32mTenha um bom dia!\033[m')
        sleep(0.5)
        exit()
```
